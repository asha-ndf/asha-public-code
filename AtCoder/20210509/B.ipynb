```python
import math
def main():

  K,N,M = map(int,input().split(' '))
  A = map(int,input().split(' '))
  B = []
  L = []
  Bap = B.append
  Lap = L.append
  for i in A:
    x = i*M/N
    y = math.floor(x)
    Lap(x - y)
    Bap(y)
  for j in range(M-sum(B)):
    p = L.index(max(L))
    B[p] = B[p] +1
    L[p]= 0
  print(*B)
main()
```
