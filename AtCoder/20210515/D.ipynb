```python
H,W = map(int,input().split(' '))
A =[]
s = 0
if H+W ==2:
  print('Draw')
  exit()
for i in range(H):
  t = input()
  for j in range(W):
    if i+j ==0:
      pass
    elif t[j] == '+':
      if (j+i)%2 ==1:
        s = s+1
      else:
        s = s-1
    else:
      if (j+i)%2 ==1:
        s = s-1
      else:
        s = s+1

if s>-1:
  print('Takahashi')
elif s<-1:
  print('Aoki')
else:
  print('Draw')
```
