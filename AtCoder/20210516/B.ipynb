```python
P1 = []
P2 = []
N = int(input())
S = input()
T = input()
ans = 0
c = 0
P3 = [0]*N
for i in range(N):
  S_T = S[i] 
  if S_T != T[i]:
    if S_T =='1':
      P1.append(i)
    else:
      P2.append(i)
if len(P1) ==len(P2):
  for i in range(len(P1)):
    P1_T = P1[i]
    P2_T = P2[i]
    if P1_T > P2_T:
      for j in range(P2_T,P1_T):
        if S[j] == '0':
          if T[j]=='0':
            P3[j] =1
    else:
      for j in range(P1_T,P2_T):
        if S[j] == '0':
          if T[j]=='0':
            P3[j] = 1
    ans = ans +1
else:
  ans =-1
ans = ans + sum(P3)
print(int(ans))
```
