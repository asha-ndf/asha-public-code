```python
N = int(input())
A = list(map(int, input().split(' ')))
B = list(map(int, input().split(' ')))
C = list(map(int, input().split(' ')))
D = []
for i in C:
  if i <=N and i>0:
    D.append(B[i-1])
A = [i for i in A if i in D]
count = 0
for i in A:
  for j in D:
    if i == j:
      count = count+1
print(count)
```
