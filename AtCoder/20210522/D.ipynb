```python
def cmb(n, r):
    if n - r < r: r = n - r
    if r == 0: return 1
    if r == 1: return n

    numerator = [n - r + k + 1 for k in range(r)]
    denominator = [k + 1 for k in range(r)]

    for p in range(2,r+1):
        pivot = denominator[p - 1]
        if pivot > 1:
            offset = (n - r) % p
            for k in range(p-1,r,p):
                numerator[k - offset] /= pivot
                denominator[k] /= pivot

    result = 1
    for k in range(r):
        if numerator[k] > 1:
            result *= int(numerator[k])

    return result
A,B,K = list(map(int, input().split(' ')))
if A*B=0:
  print('a'*A+'b'*B)
  exit()
ans = ''
for i in range(A+B):
  tmp = cmb(A+B,A) *A/(A+B)
  if tmp >=K:
    A = A-1
    ans = ans + 'A'
  else:
    B = B-1
    K = K - tmp
    ans = ans + 'A'
print(ans)
```
