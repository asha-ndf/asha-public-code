```python
N = int(input())
A = list(map(int,input().split(' ')))
max = A[0]
ans = A[0]+ max
preadd = ans
print(ans)
for i in range(N-1):
  if max<A[i+1]:
    sabun = A[i+1]-max
    max = A[i+1]
  else:
    sabun = 0
  ans = ans + preadd + sabun*(i+2) + A[i+1]
  preadd = preadd+ sabun+A[i+1]
  print(ans)
```
