```python
H,W = list(map(int,input().split(' ')))
S = []
def trans(inp):
  if inp == 'R':
    return 1
  elif inp =='B':
    return -1
  else:
    return 0
for i in range(H):
  temp = []
  for j in input():
    if j =='R':
      temp.append(1)
    elif j =='B':
      temp.append(-1)
    else:
      temp.append(0)
  S.append(temp)
k=0
for i in range(H+W-1):
  flag = 0
  tk=0
  for h in range(i+1):
    w = i-h
    if h<H and w<W:
      if S[h][w] ==0:
        tk=1
      else:
        if flag==0:
          flag = S[h][w]
        else:
          if flag != S[h][w]:
            print(0)
            exit()
  if flag==0:
    k = k+tk
ans = (2**k) %998244353 
print(ans)
    
```
