```python
N= int(input())
A = list(map(int,input().split(' ')))
B = list(map(int,input().split(' ')))
def get_index(l, x):
    if x in l:
        return l.index(x)
    else:
      print(-1)
      exit()
for i in range(N):
  A[i] = A[i]-N+i
  B[i] = B[i]-N+i

change = True
while change:
  change = False
  for i in range(N - 1):
    if B[i] > B[i + 1]:
      B[i], B[i + 1] = B[i + 1], B[i]
      A[i], A[i + 1] = A[i + 1], A[i]
      change = True
ans =0
while change:
  change = False
  for i in range(N - 1):
    if A[i] > A[i + 1]:
      A[i], A[i + 1] = A[i + 1], A[i]
      change = True
      ans = ans+1
if A==B:
  print(ans)
else:
  print(-1)
```
