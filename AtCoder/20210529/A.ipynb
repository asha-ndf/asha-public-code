```python
N = int(input())
fx,fy = list(map(int,input().split(' ')))
xmax = fx
xmin = fx
x2nd = 0
ymax = fy
ymin = fy
y2nd = 0
for i in range(N-1):
  x, y = list(map(int,input().split(' ')))
  if x>xmax:
    if x2nd < xmax-xmin:
      x2nd = xmax-xmin #古い方
    xmax=x
  elif x<xmin:
    if x2nd <xmax-xmin:
      x2nd = xmax-xmin
    xmin = x
  else:
    x2nd = max(x2nd, x-xmin, xmax-x)
  
  if y>ymax:
    if y2nd < ymax-ymin:
      y2nd = ymax-ymin #古い方
    ymax=y
  elif y<ymin:
    if y2nd <ymax-ymin:
      y2nd = ymax-ymin
    ymin = y
  else:
    y2nd = max(y2nd, y-ymin, ymax-y)
anss = [x2nd,y2nd]
anss.append(xmax-xmin)
anss.append(ymax-ymin)
print(sorted(list(anss))[-2])
      
    
```
