```python
N = int(input())
Rs = []
Gs = []
Bs = []
for i in range(N*2):
  inp = input().split(' ')
  if inp[1] =='R':
    Rs.append(int(inp[0]))
  elif inp[1] =='G':
    Gs.append(int(inp[0]))
  elif inp[1] =='B':
    Bs.append(int(inp[0]))
if (len(Rs)%2)+(len(Gs)%2)+(len(Bs)%2)==0:
  print(0)
  exit()
if len(Rs)%2 ==0:
  U1 = Gs
  U2 = Bs
elif len(Gs)%2 ==0:
  U1 = Rs
  U2 = Bs
elif len(Bs)%2 ==0:
  U1 = Gs
  U2 = Rs
min = 10**15
U1.sort()
U2.sort()
l1 = len(U1)
l2 = len(U2)
j = 0
for i in range(l1):
  flag = 1
  while (flag):
    if abs(U1[i]-U2[j]) <min:
      min = abs(U1[i]-U2[j])
    if U1[i]>U2[j]:
      if j ==l2-1:
        print(min)
        exit()
      else:
        j =j+1
    else:
      if j !=0:
        j = j-1
      flag =0 
print(min)

```
