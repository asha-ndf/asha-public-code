```python
T = int(input())
for i in range(T):
  N = int(input())
  P = list(map(int,input().split(' ')))
  T = sorted(P)
  count = 0
  ans = ''
  for j in range(N):
    for k in range(N-1):
      if P[k]>P[k+1]:
        if k%2 == count%2:
          if ans =='':
            ans = str(k+1)
          else:
            ans=ans+' '+str(k+1)
          P[k],P[k+1] = P[k+1],P[k]
      count= count+1
    if P == T:
      break
  if ans =='':
    print(0)
  else:
    print(ans)
```
