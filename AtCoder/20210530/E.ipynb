```python
N, M = list(map(int, input().split(' ')))
XY=[]
Xs=[]
for i in range(M):
  row =input() 
  XY.append(row)
  Xs.append(int(row.split(' ')[0]))
yset = set([N])
for p in range(len(Xs)-1):
  i = Xs[p+1]
  newset =set()
  for j in yset:
    if (str(i)+' '+str(j-1)) in XY:
      newset.add(j-1)
    if (str(i)+' '+str(j+1)) in XY:
      newset.add(j+1)
    if (str(i)+' '+str(j)) in XY:
      pass
    else:
      newset.add(j)
    if len(newset)==0:
      print(0)
      exit()
  yset =newset
print(len(yset))
```
