```python
import numpy as np
N,M = map(int, input().split(' '))
L = np.zeros((N, N))
for i in range(N):
  L[i][i]=1
for i in range(M):
  a,b = map(int, input().split(' '))
  L[a-1][b-1]=1
for i in range(M):
  for j in range(M):
    if L[i][j]==0:
      for p in range(M):
        if L[i][p] * L[p][j] ==1:
          L[i][j]==1
          break
print(L.sum())
```
