```python
N = int(input())
T = list(map(int, input().split(' ')))
all = sum(T)
mini = int((all+1)/2)
T.sort()
G1 = []
G2 = []
ans = 0
for i in reversed(T):
  if ans+i <=mini:
    ans = ans+i
    G1.append(i)
  else:
    G2.append(i)

ans1 = sum(G1)
ans2 = sum(G2)
temp = 0
if ans1>ans2:
  if G1[0]>G2[0]:
    for i in G2:
      if temp+i <G1[0]:
        temp = temp+i
      else:
        oans1 = ans1-G1[0]+temp+i
        oans2 = ans2-temp+G1[0]-i
  else:
    for i in G1:
      if temp+i <G2[0]:
        temp = temp+i
      else:
        oans1 = ans1+G2[0]-temp
        oans2 = ans2-G2[0]+temp
else:
  if G1[0]>G2[0]:
    for i in G2:
      if temp+i <G1[0]:
        temp = temp+i
      else:
        oans1 = ans1-G1[0]+temp
        oans2 = ans2-temp+G1[0]
  else:
    for i in G1:
      if temp+i <G2[0]:
        temp = temp+i
      else:
        oans1 = ans1+G2[0]-temp-i
        oans2 = ans2-G2[0]+temp+i
a1 = int(max(ans1,ans2))
a2 = int(max(oans1,oans2))

print(min(a1,a2))
```
